```python
#seaborn parametrlari:
sns.set_theme(
    context="notebook", # o'lcham konteksti
    style="darkgrid", # fon stili
    palette="deep", # rang palitasi
    font_scale=1.2 # shrift kattaligi
    )
```

```python
from matplotlib import axes
#axes_style()
with sns.axes_style("white"): #faqat bir martalik sozlash
  sns.histplot(data=tips, x="total_bill", kde=True)
  plt.show()

sns.set_style("whitegrid", {
    "grid.linestyle": "--",
    "grid.linewidth": 0.8,
    "axes.spines.top": False,
    "axes.spines.right": False,
})


sns.histplot(data=tips, x="tip")
plt.show()
```

```python
#sns.set_style() - grid chizig'i yo'g'onligi, shrift va boshqalar
sns.set_style("whitegrid", {
    "grid.linestyle": "--",
    "grid.linewidth": 0.8,
    "axes.spines.top": False,
    "axes.spines.right": False
})
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", alpha= 0.8, s =200, c = "#dbb62e", hue = "day", markers = ['o','D','^','s'])
plt.show()
```

```python
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="day",
    style="day",
    markers=['o', 'D', '^', 's'],
    s=200,
    alpha=0.8
)
```

```python
plt.figure(figsize = (10,6))
sns.scatterplot(
    data = tips,
    x="total_bill",
    y = "tip",
    # hue="sex",
    s=100,
    alpha = 0.7,
    # size = "size",
    style = 'smoker',
    marker = 'D'
)
plt.show()
```

```python
sns.kdeplot(data=tips, x="total_bill", hue = "day", color = 'steelblue', linestyle="--", linewidth=0.2, fill=True, bw_adjust=0.05)
plt.show()
```

```python
#fill - to'ldirish, ichini bo'yab beradi,
#bw_adjust - curve qanchalik smooth yoki detailed bo‘lishini belgilaydi.
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex", width = 0.4,gap=0.5)
plt.legend(
    title = 'salom',
    fontsize = 8,
    title_fontsize = 10,
    markerscale = 10,
    loc = "upper left"
)

plt.show()
```

```python
sns.stripplot(data=tips,
              x ="day",
              y = "total_bill",
              jitter = True,
              alpha =0.5,
              marker = 'o',
              s = 7,
              color = 'red'
              )
plt.show()
```

```python
sns.boxplot(data=tips, x= 'day', y='total_bill', width = 0.4)
sns.swarmplot(data=tips, x= 'day', y='total_bill', color = 'black', size = 3)
plt.show()
```

```python
sns.regplot(
 data=tips,
 x="total_bill",
 y="tip",
 ci=95,
 # 95% ishonch oralig'i (default)
 # ci=68 # torroq
 # ci=None # o'chirish
)
plt.show()
```

```python
sns.jointplot(
 data=tips,
 x="total_bill",
 y="tip",
 kind="kde",
 fill = True,
)
plt.show()
```
